```python
num_rating=rating_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
```

```python
avg_rating=rating_with_name.groupby('Book-Title').mean()['Book-Rating'].reset_index()
avg_rating.rename(columns={'Book-Rating':'Avg-Rating'},inplace=True)
avg_rating
```

```python
popularity_df=num_rating.merge(avg_rating,on='Book-Title')
```

```python
popularity_df
```

```python
popularity_df[popularity_df['Num-Rating']>=250]
```

```python
popularity_df=popularity_df[popularity_df['Num-Rating']>=250].sort_values('Avg-Rating',ascending=False).head(50)
```

```python
popularity_df.merge(book,on='Book-Title')
```

```python
popularity_df.merge(book,on='Book-Title').drop_duplicates('Book-Title')
```

```python
popular_df=popularity_df.merge(book,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','Num-Rating','Avg-Rating']]
```

# Collaborative Filtering Based Recommender

```python
#collecting the users who have rated more than 200 books
```

```python
x=rating_with_name.groupby('User-ID').count()['Book-Rating']>200
educated_user = x[x].index
```

```python
filtered_rating = rating_with_name[rating_with_name['User-ID'].isin(educated_user)]
```

```python
filtered_rating
```

```python
#collecting data of the books which got at least 50 rating
```

```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating'] >= 50
famous_book = y[y].index
```

```python
final_rating = filtered_rating[filtered_rating['Book-Title'].isin(famous_book)]
```

```python
final_rating.drop_duplicates()
```

```python
pivot_table = final_rating.pivot_table(index='Book-Title', columns='User-ID',values='Book-Rating')
pivot_table.fillna(0,inplace=True)
```

```python
pivot_table
```

```python
#finding the distance of everybook from other.
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(pivot_table)
```

```python
similarity_score.shape
```

```python
similarity_score
```

```python
def recommend(book_name):
    #fetching the index
    index = np.where(pivot_table.index==book_name)[0][0]
    similar_item = sorted(list(enumerate(similarity_score[index])),key=lambda x:x[1],reverse=True)[1:6]
    
    for i in similar_item:
        print(pivot_table.index[i[0]])
```

```python
recommend('A Bend in the Road')
```

```python
def recommend(book_name):
    #fetching the index
    index = np.where(pivot_table.index==book_name)[0][0]
    similar_item = sorted(list(enumerate(similarity_score[index])),key=lambda x:x[1],reverse=True)[1:6]
    
    data=[]
    for i in similar_item:
        item=[]
        temp_df=book[book['Book-Title'] == pivot_table.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title']))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author']))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M']))
        #item.extend(list(temp_df.drop_duplicates('Book-Title')['Avg-Rating']))
        
        data.append(item)
        
    return data
        
```
